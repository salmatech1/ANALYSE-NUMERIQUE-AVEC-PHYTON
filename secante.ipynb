```python
def secante(f, x0, x1, eps=1e-6, max_iter=100):
    """
    Méthode de la sécante pour la recherche de racine de f.
    """
    for k in range(max_iter):
        if f(x1) == f(x0):
            raise ValueError("Division par zéro : f(x1) = f(x0).")
        x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        if abs(x2 - x1) < eps:
            return x2, k
        x0, x1 = x1, x2
    return x2, max_iter
```

```python
f = lambda x: x**2 - 2  
racine, k = secante(f, 1, 2, eps=1e-11, max_iter=100)
print(f"Racine ≈ {racine:.10f} trouvée en {k} itérations.")
```
