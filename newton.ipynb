```python
def newton(f, df, x0, eps=1e-6, max_iter=100):
    """
    Méthode de Newton pour la recherche de racine.
    """
    for k in range(max_iter):
        if df(x0) == 0:
            raise ValueError("Dérivée nulle : la méthode ne peut pas continuer.")
        x1 = x0 - f(x0) / df(x0)
        if abs(x1 - x0) < eps:
            return x1, k
        x0 = x1
    return x1, max_iter
```

```python
f = lambda x: x**2 - 2
df = lambda x: 2*x
racine, k = newton(f, df, 1, eps=1e-11, max_iter=100)
print(f"Racine ≈ {racine:.10f} trouvée en {k} itérations.")
```
